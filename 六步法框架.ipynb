````python
# NVIDIA未来增长率预测分析 - 基于yfinance真实数据
# =====================================================

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
import yfinance as yf

# 配置中文字体
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS', 'SimHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

print("📈 NVIDIA 基于真实财务数据的增长趋势分析与未来预测")
print("=" * 60)

# 1. 从yfinance获取真实财务数据
ticker = "NVDA"
nvda = yf.Ticker(ticker)

try:
    # 获取年度财务数据
    financials = nvda.financials

    if financials is not None and len(financials) > 0:
        # 提取总收入和净利润数据
        total_revenue = financials.loc['Total Revenue']
        net_income = financials.loc['Net Income']

        # 按年份排序 (从早到晚)
        total_revenue = total_revenue.sort_index()
        net_income = net_income.sort_index()

        print("📊 获取到的历史财务数据:")
        print("-" * 35)

        # 显示原始数据
        for date, revenue, income in zip(total_revenue.index, total_revenue.values, net_income.values):
            fiscal_year = date.strftime('%Y')
            print(f"{fiscal_year}: 收入 ${revenue/1e9:.1f}B, 净利润 ${income/1e9:.1f}B")

        # 2. 计算年增长率
        revenue_growth_rates = total_revenue.pct_change().dropna() * 100
        profit_growth_rates = net_income.pct_change().dropna() * 100

        print(f"\n📈 历史增长率分析:")
        print("-" * 25)

        growth_data = []
        for date, rev_growth, profit_growth in zip(revenue_growth_rates.index,
                                                   revenue_growth_rates.values,
                                                   profit_growth_rates.values):
            fiscal_year = date.strftime('%Y')
            print(f"{fiscal_year}: 收入增长 {rev_growth:+.1f}%, 净利润增长 {profit_growth:+.1f}%")
            growth_data.append({
                'year': int(fiscal_year),
                'revenue_growth': rev_growth,
                'profit_growth': profit_growth
            })

        # 转换为数组进行分析
        growth_df = pd.DataFrame(growth_data)
        years = growth_df['year'].values
        revenue_growth = growth_df['revenue_growth'].values
        profit_growth = growth_df['profit_growth'].values

        # 3. 趋势分析和预测模型
        def predict_growth_trend(years_data, growth_rates, method='smart'):
            """
            基于历史数据预测增长趋势
            method: 'linear', 'polynomial', 'smart' (综合判断)
            """
            years_array = np.array(years_data)
            growth_array = np.array(growth_rates)

            if len(growth_array) < 2:
                return 0, 0.1

            if method == 'linear':
                # 线性回归预测
                slope, intercept, r_value, p_value, std_err = stats.linregress(years_array, growth_array)
                next_year_pred = slope * (years_array[-1] + 1) + intercept
                confidence = abs(r_value)

            elif method == 'polynomial' and len(growth_array) >= 3:
                # 二次多项式拟合
                coeffs = np.polyfit(years_array, growth_array, min(2, len(growth_array)-1))
                poly = np.poly1d(coeffs)
                next_year_pred = poly(years_array[-1] + 1)
                # 计算R方值
                predicted = poly(years_array)
                ss_res = np.sum((growth_array - predicted) ** 2)
                ss_tot = np.sum((growth_array - np.mean(growth_array)) ** 2)
                confidence = max(0, 1 - (ss_res / ss_tot))

            else:  # smart method - 综合判断
                # 分析最近的趋势
                if len(growth_array) >= 2:
                    recent_trend = growth_array[-1] - growth_array[-2]

                    # 如果最近在改善
                    if recent_trend > 0:
                        # 预测继续改善，但速度放缓
                        improvement_factor = min(0.8, abs(recent_trend) / 100)  # 防止过度乐观
                        next_year_pred = growth_array[-1] + recent_trend * improvement_factor
                        confidence = 0.6
                    else:
                        # 如果还在下滑，预测小幅改善或持平
                        next_year_pred = growth_array[-1] + abs(growth_array[-1]) * 0.1
                        confidence = 0.4
                else:
                    next_year_pred = growth_array[-1]
                    confidence = 0.3

            return next_year_pred, confidence

        # 4. 多种方法预测下一年增长率
        next_year = years[-1] + 1

        print(f"\n🔮 {next_year}年增长率预测 (基于真实数据):")
        print("-" * 40)

        # 收入增长预测
        rev_linear, rev_conf_linear = predict_growth_trend(years, revenue_growth, 'linear')
        if len(revenue_growth) >= 3:
            rev_poly, rev_conf_poly = predict_growth_trend(years, revenue_growth, 'polynomial')
        else:
            rev_poly, rev_conf_poly = rev_linear, rev_conf_linear
        rev_smart, rev_conf_smart = predict_growth_trend(years, revenue_growth, 'smart')

        print("📈 收入增长率预测:")
        print(f"  • 线性趋势法: {rev_linear:+.1f}% (置信度: {rev_conf_linear:.2f})")
        if len(revenue_growth) >= 3:
            print(f"  • 多项式拟合: {rev_poly:+.1f}% (置信度: {rev_conf_poly:.2f})")
        print(f"  • 智能预测: {rev_smart:+.1f}% (置信度: {rev_conf_smart:.2f})")

        # 净利润增长预测
        profit_linear, profit_conf_linear = predict_growth_trend(years, profit_growth, 'linear')
        if len(profit_growth) >= 3:
            profit_poly, profit_conf_poly = predict_growth_trend(years, profit_growth, 'polynomial')
        else:
            profit_poly, profit_conf_poly = profit_linear, profit_conf_linear
        profit_smart, profit_conf_smart = predict_growth_trend(years, profit_growth, 'smart')

        print(f"\n💰 净利润增长率预测:")
        print(f"  • 线性趋势法: {profit_linear:+.1f}% (置信度: {profit_conf_linear:.2f})")
        if len(profit_growth) >= 3:
            print(f"  • 多项式拟合: {profit_poly:+.1f}% (置信度: {profit_conf_poly:.2f})")
        print(f"  • 智能预测: {profit_smart:+.1f}% (置信度: {profit_conf_smart:.2f})")

        # 5. 可视化分析
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

        # 收入增长趋势图
        ax1.plot(years, revenue_growth, 'bo-', linewidth=2, markersize=8, label='Historical Data')

        # 添加预测点
        ax1.plot(next_year, rev_smart, 'rs', markersize=10,
                label=f'{next_year} Prediction: {rev_smart:+.1f}%')

        # 添加趋势线
        if len(years) >= 2:
            z = np.polyfit(years, revenue_growth, 1)
            p = np.poly1d(z)
            extended_years = np.array(list(years) + [next_year])
            ax1.plot(extended_years, p(extended_years), 'r--', alpha=0.5, label='Trend Line')

        ax1.set_title('Revenue Growth Rate Trend & Prediction', fontsize=12, fontweight='bold')
        ax1.set_xlabel('Fiscal Year')
        ax1.set_ylabel('Revenue Growth Rate (%)')
        ax1.grid(True, alpha=0.3)
        ax1.legend()
        ax1.axhline(y=0, color='black', linestyle='-', alpha=0.3)

        # 净利润增长趋势图
        ax2.plot(years, profit_growth, 'go-', linewidth=2, markersize=8, label='Historical Data')

        # 添加预测点
        ax2.plot(next_year, profit_smart, 'rs', markersize=10,
                label=f'{next_year} Prediction: {profit_smart:+.1f}%')

        # 添加趋势线
        if len(years) >= 2:
            z = np.polyfit(years, profit_growth, 1)
            p = np.poly1d(z)
            ax2.plot(extended_years, p(extended_years), 'r--', alpha=0.5, label='Trend Line')

        ax2.set_title('Net Income Growth Rate Trend & Prediction', fontsize=12, fontweight='bold')
        ax2.set_xlabel('Fiscal Year')
        ax2.set_ylabel('Net Income Growth Rate (%)')
        ax2.grid(True, alpha=0.3)
        ax2.legend()
        ax2.axhline(y=0, color='black', linestyle='-', alpha=0.3)

        plt.tight_layout()
        plt.show()

        # 6. 综合分析与投资建议
        print(f"\n🎯 基于真实数据的投资分析:")
        print("-" * 35)

        expected_revenue_growth = rev_smart
        expected_profit_growth = profit_smart

        print(f"💡 {next_year}年最可能情形:")
        print(f"  • 预期收入增长: {expected_revenue_growth:+.1f}%")
        print(f"  • 预期净利润增长: {expected_profit_growth:+.1f}%")

        # 基于预测结果给出投资建议
        if expected_revenue_growth > 20 and expected_profit_growth > 20:
            outlook = "📈 强劲增长"
            recommendation = "积极买入，长期持有"
        elif expected_revenue_growth > 0 and expected_profit_growth > 0:
            outlook = "🔄 温和复苏"
            recommendation = "逢低买入，分批建仓"
        elif expected_revenue_growth > revenue_growth[-1]:
            outlook = "🔄 缓慢改善"
            recommendation = "谨慎观察，等待更明确信号"
        else:
            outlook = "⚠️  持续挑战"
            recommendation = "建议观望，关注基本面转折点"

        print(f"\n📊 投资展望: {outlook}")
        print(f"💰 建议策略: {recommendation}")

        # 计算一些额外的有用指标
        avg_revenue_growth = np.mean(revenue_growth)
        avg_profit_growth = np.mean(profit_growth)
        revenue_volatility = np.std(revenue_growth)
        profit_volatility = np.std(profit_growth)

        print(f"\n📊 历史表现统计:")
        print(f"  • 平均收入增长: {avg_revenue_growth:+.1f}%")
        print(f"  • 平均净利润增长: {avg_profit_growth:+.1f}%")
        print(f"  • 收入增长波动性: {revenue_volatility:.1f}%")
        print(f"  • 净利润增长波动性: {profit_volatility:.1f}%")

    else:
        print("❌ 无法获取财务数据，可能是数据源问题")

except Exception as e:
    print(f"❌ 获取数据时出错: {e}")
    print("建议检查网络连接或稍后重试")

print(f"\n⚠️  重要提示:")
print("  • 预测基于历史数据，仅供参考")
print("  • 实际结果可能因市场变化而异")
print("  • 建议结合最新财报和行业动态")
print("  • 投资有风险，决策需谨慎")
````

````python
# 第三步：处理数据 (Process Data) - 阿里巴巴财务数据处理
def safe_format_baba(value):
    """安全格式化数值，处理None值"""
    if value is None or pd.isna(value):
        return "N/A"
    if isinstance(value, (int, float)):
        return f"{value:,.0f}"
    return str(value)

print("=== 阿里巴巴财务数据处理 ===")

try:
    # 获取总收入数据
    total_revenue_baba = financials_baba.loc['Total Revenue'] if 'Total Revenue' in financials_baba.index else None
    if total_revenue_baba is None:
        total_revenue_baba = financials_baba.loc['Operating Revenue'] if 'Operating Revenue' in financials_baba.index else financials_baba.iloc[0]

    # 获取净利润数据
    net_income_baba = income_baba.loc['Net Income'] if 'Net Income' in income_baba.index else None
    if net_income_baba is None:
        net_income_baba = income_baba.loc['Net Income Common Stockholders'] if 'Net Income Common Stockholders' in income_baba.index else income_baba.iloc[-1]

    # 数据预处理
    revenue_data_baba = total_revenue_baba.dropna().sort_index()
    profit_data_baba = net_income_baba.dropna().sort_index()

    print("✓ 收入数据处理完成")
    print(f"  可用年份: {len(revenue_data_baba)} 年")
    print(f"  时间范围: {revenue_data_baba.index.min()} - {revenue_data_baba.index.max()}")

    print("\n✓ 利润数据处理完成")
    print(f"  可用年份: {len(profit_data_baba)} 年")
    print(f"  时间范围: {profit_data_baba.index.min()} - {profit_data_baba.index.max()}")

    # 显示最近3年财务概况
    print(f"\n=== 阿里巴巴最近3年财务概况 ===")
    recent_years = revenue_data_baba.index[-3:]
    for year in recent_years:
        revenue_val = revenue_data_baba[year] if year in revenue_data_baba.index else None
        profit_val = profit_data_baba[year] if year in profit_data_baba.index else None

        print(f"{year}年:")
        print(f"  总收入: ¥{safe_format_baba(revenue_val/1e8)} 亿元")
        print(f"  净利润: ¥{safe_format_baba(profit_val/1e8)} 亿元")

        if revenue_val and profit_val and revenue_val != 0:
            margin = (profit_val / revenue_val) * 100
            print(f"  净利率: {margin:.1f}%")
        print()

except Exception as e:
    print(f"❌ 数据处理失败: {e}")
    print("将使用模拟数据进行演示...")
````

````python
# 第四步：分析和解释 (Analyze and Interpret) - 阿里巴巴趋势分析
print("=== 阿里巴巴财务趋势分析 ===")

try:
    # 计算增长率
    if len(revenue_data_baba) >= 2:
        revenue_growth_rates_baba = revenue_data_baba.pct_change().dropna()
        profit_growth_rates_baba = profit_data_baba.pct_change().dropna()

        print("📈 收入增长率分析:")
        for year in revenue_growth_rates_baba.index:
            growth = revenue_growth_rates_baba[year] * 100
            print(f"  {year}: {growth:+.1f}%")

        print(f"\n📈 利润增长率分析:")
        for year in profit_growth_rates_baba.index:
            growth = profit_growth_rates_baba[year] * 100
            print(f"  {year}: {growth:+.1f}%")

        # 平均增长率
        avg_revenue_growth_baba = revenue_growth_rates_baba.mean()
        avg_profit_growth_baba = profit_growth_rates_baba.mean()

        print(f"\n📊 平均增长率:")
        print(f"  收入平均增长率: {avg_revenue_growth_baba*100:+.1f}%")
        print(f"  利润平均增长率: {avg_profit_growth_baba*100:+.1f}%")

        # 波动性分析
        revenue_volatility_baba = revenue_growth_rates_baba.std()
        profit_volatility_baba = profit_growth_rates_baba.std()

        print(f"\n📊 增长稳定性分析:")
        print(f"  收入增长波动率: {revenue_volatility_baba*100:.1f}%")
        print(f"  利润增长波动率: {profit_volatility_baba*100:.1f}%")

        # 财务健康度评估
        print(f"\n🏥 财务健康度评估:")

        # 收入增长稳定性
        if avg_revenue_growth_baba > 0:
            if revenue_volatility_baba < 0.2:
                revenue_health = "收入增长稳定且健康"
            else:
                revenue_health = "收入增长但波动较大"
        else:
            revenue_health = "收入面临下滑压力"

        # 利润增长稳定性
        if avg_profit_growth_baba > 0:
            if profit_volatility_baba < 0.3:
                profit_health = "利润增长稳定"
            else:
                profit_health = "利润增长但不稳定"
        else:
            profit_health = "利润面临挑战"

        print(f"  {revenue_health}")
        print(f"  {profit_health}")

    else:
        print("⚠️ 数据不足，无法进行趋势分析")

except Exception as e:
    print(f"❌ 分析过程出错: {e}")
    print("使用备用分析方法...")
````
